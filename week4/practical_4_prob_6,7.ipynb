```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("colwidth", 12)
```

```python
#6
df = pd.read_csv("netflix_titles.csv")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"]).dt.to_period("M")
```

```python
df.sort_values(by = "date_added", inplace = True)
```

```python
df["date_added"].value_counts().plot()                
```

```python
df.groupby("date_added").count().iloc[:,0].plot()
```

```python
df.groupby("date_added").count().iloc[:,0].plot(kind = "hist")
```
