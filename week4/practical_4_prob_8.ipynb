```python
import pandas as pd
pd.set_option('max_colwidth',14)
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head(3)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
df.sort_values(by = "date_added", inplace = True)
```

```python
df['diffs'] = df.groupby("director")['date_added'].diff()
```

```python
df["diffs"] = df["diffs"].dt.days
```

```python
df["diffs"] = df["diffs"].fillna("-1")
```
