```python
import pandas as pd
pd.set_option("max_colwidth", 12)
```

```python
#4
df = pd.read_csv("netflix_titles.csv")
```

```python
df["cast"] = df["cast"].astype("str").str.split(",")
```

```python
df.explode("cast")
```

```python
#5
import matplotlib.pyplot as plt
def duration_fix(x):
    return x.split()[0]

df["duration"] = df["duration"].apply(duration_fix)

df["date_added"] = pd.to_datetime(df["date_added"])

df = df[df["cast"].astype("str").str.contains("Antonio Banderas")].sort_values(by = "date_added")

x = df["date_added"]
y = df["duration"]

plt.plot(x,y)
```
